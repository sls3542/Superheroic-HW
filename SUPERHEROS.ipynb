Compare the average weight of super powers who have Super Speed to those who do not.

```python
speedweight = merged.loc[:,['Hero','Weight','Super Speed']]
speedweight['Super Speed'] = speedweight['Super Speed'].map({True: 'yes', False: 'no'})
speedweight.head()
```

```python
speedweight['Weight'] = speedweight['Weight'].str.split(' ').str[0]
speedweight['Weight'].head()
```

```python
speedweight['Weight'] = pd.to_numeric(speedweight['Weight'])
```

```python
avg_weight = speedweight.groupby('Super Speed')['Weight'].mean().reset_index()
avg_weight.rename(columns={"Weight": "Average Weight"},inplace = True)
avg_weight
```
